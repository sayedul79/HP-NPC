### Import library

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
#%matplotlib inline
print('All essential libraries are imported')
```

### Find the value of a, b, c, d and e for which the following relationship is satisfied $$R=a \theta^{4}+b \theta^{3}+c \theta^{2}+d \theta+e$$

```python
T=np.array([90, 95, 100, 105, 110, 115,
           120, 125, 130])
R=np.array([279, 285, 291, 299, 308, 317,
           325, 333, 343])
print("Temparature(Celcius) :", T)
print('Resistance(Ohm) :', R)
```

```python
curve4=np.polyfit(T, R, 4)
curve4
```

```python
poly4=np.poly1d(curve4)
print(poly4)
```

```python
plt.figure(1)

x=np.linspace(0, 200, 200)
plt.scatter(T, R)
plt.plot(x, poly4(x), 'r')

plt.title(
    "Curve fitting on $a \Theta^{4}+b \Theta^{3}+c \Theta^{2}+d \Theta+e$")
plt.xlabel('Temparature($\Theta$ in $^{\circ}$C)')
plt.ylabel('Resistance(R in $\Omega$)')
```
