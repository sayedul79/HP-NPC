# Line Plot

```python
#%matplotlib notebook
import numpy as np
from matplotlib import pyplot as plt
#print(plt.style.available)
#plt.style.use('fivethirtyeight')

def f(t):
    return t ** 2 * np.exp(-t ** 2)


t= np.linspace(0, 3, 51)  # 51 points between 0 and 3
y = np.zeros(len(t))  # allocate y with float elements
for i in np.arange(len(t)):
    y[i] = f(t[i])
plt.plot(t, y, label='t^2*exp(-t^2)', color='r')
plt.xlabel('t')
plt.ylabel('y')
plt.axis([0, 3, -0.05, 0.6])
plt.legend()   # [tmin, tmax, ymin, ymax]
plt.title('My First Matplotlib Demo')
plt.show()
```

# Multiple Subplot

```python
import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

# log plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for plotting
t = np.arange(0.01, 20.0, 0.01)
plt.semilogy(t,np.exp(-t / 5.0))
plt.grid()
```

```python
plt.semilogx(t,np.sin(2*np.pi*t))
plt.grid()
```

```python
plt.loglog(t,20 * np.exp(-t / 10.0))
plt.grid()
```
